## 1. What is the total revenue generated by Olist, and how has it changed over time?

#### Clean the orders dataframe

```python
orders.head()
```

```python
# 1. Create a new column with just the date
orders['order_purchase_date'] = orders['order_purchase_timestamp'].dt.date

# 2. Convert to datetime (optional, only if you want datetime format)
orders['order_purchase_date'] = pd.to_datetime(orders['order_purchase_date'])
# 3. Drop unnecessary columns
orders2 = orders.drop([ 'order_purchase_timestamp','order_approved_at', 'order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date'], axis=1)

orders2.head()
```

```python
orders_payments.drop(orders_payments[orders_payments['order_status'] == 'canceled'].index, axis = 0 , inplace = True)
orders_payments.drop(orders_payments[orders_payments['order_status'] == 'unavailable'].index, axis = 0, inplace = True)
```

## a) Find the total revenue

```python
total_revenue = orders_payments['payment_value'].sum()
total_Revenue_Millions = round(total_revenue/1000000,2)

print('The total revenue generated by Olist is R${}M'.format(total_Revenue_Millions))
```

## b)Find the trend of revenue overtime

```python
#1. create a new dataframe with only the order purchase date and payment value columns 
revenue_trend = orders_payments.loc[:,['order_purchase_date', 'payment_value']]
#2. set the order_purchase_date as the index of the dataframe
revenue_trend = revenue_trend.set_index('order_purchase_date')
#3. Group the order_puchase_date by quarter and find the sum of payment value for each quarter
quarter_total_revenue = revenue_trend.resample('Q', closed='right').agg({'payment_value': 'sum'})
# Show result
quarter_total_revenue
```

```python
#Line plot for Quaterly Revenue
quarter_total_revenue.plot(figsize=(10,5))
plt.title('Quarterly Revenue line plot')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```

According, to the line plot above, the 3rd Quarter(Sep-Dec) of 2016 showed the  lowest revenue, with total payment_value of 136.23 while the 2nd Quarter(Apr-Aug) of 2018 showed the highest revenue with total payment_value of 3235342.38.

# 7. How many customers have made repeat purchases on Olist, and what percentage of total sales do they account for?

```python
customers.head()
```

#### a)How many customers have made repeat purchases on Olist

```python
unique_customer_id = len(customers['customer_id'].unique())
unique_customer_id
```

```python
customer_unique_id = len(customers['customer_unique_id'].unique())
customer_unique_id
```

```python
uniquecustomers = customers['customer_unique_id'].value_counts()
uniquecustomers = uniquecustomers[uniquecustomers > 1]
repeating_customers = len(uniquecustomers)
repeating_customers
```

#### b) What percentage of total sales do the repeating customers account for?

We need to join the customers,orders and items table

```python
cust_order = pd.merge(customers,orders , on ='customer_id',how='inner')
ico_join = pd.merge(cust_order,items, on ='order_id',how='inner')
ico_join.head(5)
```

##### Drop unnecesary rows 

```python
ico_join.drop(ico_join[ico_join['order_status']=='canceled'].index, axis=0,inplace=True)
ico_join.drop(ico_join[ico_join['order_status']=='unavailable'].index,axis=0,inplace=True)
```

```python
#Group orders by customer_unique_id and count the number of orders for each customer
customer_order = ico_join.groupby('customer_unique_id')['order_id'].count()
#Filter customers with more than one order and sum up their sales
repeating_customers = customer_order[customer_order>1]
# Find the sum of the price for all customers that have placed more than 1 order
total_price = ico_join[ico_join['customer_unique_id'].isin(repeating_customers.index)].groupby('customer_unique_id')['price'].sum().sum()
# Find the sum of the freight value for all customers that have placed more than 1 order
total_freight = ico_join[ico_join['customer_unique_id'].isin(repeating_customers.index)].groupby('customer_unique_id')['price'].sum().sum()
#Find the sum of both price and freight value
total_sales = total_price + total_freight
# Find the percentage of total sales made by repeating customer is
per_total_sales = round((total_sales/total_revenue)*100,2)
print('The percentage of  repeating customers is:',per_total_sales)
```

⬩ Continue prioritizing credit card payments, as they yield higher transaction values.

⬩ Explore installment options or promotions for high-value categories like computers & accessories to further increase average order value.

### 5. Reduce Order Cancellations

⬩ Investigate root causes of cancellations (delivery delays, stock issues, seller reliability).

⬩ Introduce stricter seller performance metrics and clearer delivery timelines.

⬩ Even a small reduction in cancellations could recover lost revenue and improve customer trust.

⬩ Maintain dominance in São Paulo through faster deliveries and premium services.
